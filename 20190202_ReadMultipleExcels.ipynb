```python
import pandas as pd
%matplotlib inline
```

```python
url = 'https://github.com/pongsakcu/Dataset/blob/master/Car1_whole2559.xls?raw=true'
```

## read a single sheet

```python
colnames = ['brand','1300cc','1600cc','1800cc','2000cc','gt2000cc','elec','total']
df=pd.read_excel(url, sheet_name=1, usecols='B:I', skiprows=7, skipfooter=3, header=None, names=colnames)
df.head()
```

## read multiple sheets

```python
colnames = ['brand','1300cc','1600cc','1800cc','2000cc','gt2000cc','elec','total']
dfs = [] # list of dataframes
dfs = [pd.read_excel(url, sheet_name=sn, usecols='B:I', skiprows=7, skipfooter=3, header=None, names=colnames)
#         for sn in range(0,12)]  # from sheet 'Total' to 'December'
         for sn in range(1,13)]  # from sheet 'January' to 'December'
```

```python
dfs[1].head()
```

```python
dfs[11].head()
```

```python
[d.set_index('brand', inplace=True) for d in dfs]
```

```python
dfs[0]
```

## concat = aggregation

```python
df = pd.concat(dfs, keys=range(1,13), names=['month', 'brand'])
df.sample(10)
```

```python
df.head(100)
```

```python
df = df[df.total !=0]
```

```python
df.sample(n=10)
```

```python
df.reset_index(inplace=True)
#df.sample(n=10)
df.head(100)
```

## save to csv

```python
df.to_csv('car1_2559.csv', index=False)
```

## save multiple dataframes to excel

```python
sheets = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
excel_file = pd.ExcelWriter('CarSheet.xlsx')
[c.to_excel(excel_file, sheet_name=sheets[i]) for i, c in enumerate(dfs)]
excel_file.close()
```
