```python
# อ่าน excel กำหนดชีท ตัดขอบหน้า ขอบหลัง ตัดหัว ตัดท้าย เปลี่ยนชื่อ column
colnames = ['brand','1300cc','1600cc','1800cc','2000cc','gt2000cc','elec','total']
df = pd.read_excel(path+'Car1_whole2559.xls', sheet_name='Total', usecols='B:I', skiprows=6, skipfooter=3, names=colnames)
df
```

```python
# ให้ brand เป็น index
df.set_index('brand', inplace=True)
df
```

```python
# ตัดแถวที่มีค่า = 0
df = df[df.total !=0]
df
```

```python
# top 10
df.nlargest(10, 'total')
```

```python
# plot data top 10
df.nlargest(10, 'total')['total'].plot(kind='barh')
```

```python
# plot data top 10 เรียงค่า
df.nlargest(10, 'total')['total'].sort_values().plot(kind='barh')
```
